```python
import pandas as pd

df = pd.read_csv(r'C:\Users\clint\Desktop\Remitances\Data\29.csv')
df
```

```python
# Filter for Kenya as receiving country in 2024
kenya_2024 = df[(df['Receiving_Country'] == 'Kenya')]


# Change United States to Kenya value to 2,637,772.27
kenya_2024.loc[kenya_2024['Sending_Country'] == 'United States', 'Value'] = "2,637,772.27"
# Display all entries
print(f"Total number of entries: {len(kenya_2024)}")
print(f"\nAll Value entries for Kenya in 2024:\n")
print(kenya_2024[['Sending_Country', 'Receiving_Country', 'Year', 'Value','Source']].to_string())

# Convert Value column to numeric (removing commas) and calculate sum
kenya_2024['Value_numeric'] = kenya_2024['Value'].astype(str).str.replace(',', '').astype(float)


total_sum = kenya_2024['Value_numeric'].sum()

print(f"\n\nTotal sum of all Value entries: {total_sum:,.2f} USD millions")
```

```python
# Check unique sending countries
print("Unique sending countries in kenya_2024:")
print(kenya_2024['Sending_Country'].unique())
```

```python
# Filter for North American countries only
north_american_countries = ['United States', 'Canada', 'Mexico']

kenya_2024_north_america = kenya_2024[kenya_2024['Sending_Country'].isin(north_american_countries)]

print(f"Total entries from North America to Kenya: {len(kenya_2024_north_america)}")
print(f"\nRemittances from North America to Kenya:\n")
print(kenya_2024_north_america[['Sending_Country', 'Receiving_Country', 'Year', 'Value', 'Source']].to_string())

# Calculate total from North America
total_north_america = kenya_2024_north_america['Value_numeric'].sum()
print(f"\n\nTotal remittances from North America: {total_north_america:,.2f} USD millions")
```

```python
diaspora_df = pd.read_csv(r'C:\Users\clint\Desktop\Remitances\Data\Diaspora Remittances.csv')
diaspora_df
```

```python
#isolate Year column 2024
diaspora_2024 = diaspora_df[diaspora_df['Year'] == 2024]
diaspora_2024
```

```python
# Convert string values to numeric (remove commas) and convert from thousands to millions
for col in ['North America', 'Europe', 'Rest of World', 'Total Remittances (USD \'000)']:
    diaspora_2024[col] = diaspora_2024[col].astype(str).str.replace(',', '').astype(float) / 1000

# Sum each value column
print("Sum of value columns for 2024 (in USD millions):\n")
print(f"North America: {diaspora_2024['North America'].sum():,.2f}")
print(f"Europe: {diaspora_2024['Europe'].sum():,.2f}")
print(f"Rest of World: {diaspora_2024['Rest of World'].sum():,.2f}")
print(f"Total Remittances (USD millions): {diaspora_2024['Total Remittances (USD \'000)'].sum():,.2f}")
```
